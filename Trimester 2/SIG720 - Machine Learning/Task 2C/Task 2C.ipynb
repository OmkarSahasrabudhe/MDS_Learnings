```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
df
```

```python
df.info()
```

The dataset contains 2111 entries and 17 columns. There are no missing values. The columns Age, Height, and Weight are already of type float64, and thus suitable for our calculations.

Given that the columns Gender, family_history_with_overweight, FAVC, CAEC, SMOKE, SCC, CALC, MTRANS, and NObeyesdad are all categorical, we will convert them into numerical representations using one-hot encoding.

We will then proceed with finding the optimal k value using the Silhouette Coefficient and plot the results.

```python
import altair as alt

# Create a variable `columns_to_encode` with the columns `Gender`, `family_history_with_overweight`, `FAVC`, `CAEC`, `SMOKE`, `SCC`, `CALC`, `MTRANS`, and `NObeyesdad`.
columns_to_encode = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad']

# One-hot encode the columns in `columns_to_encode` and concatenate the result with the original columns that are not in `columns_to_encode`.
df_encoded = pd.concat([df.drop(columns=columns_to_encode), pd.get_dummies(df[columns_to_encode], drop_first=True)], axis=1)

# Create an empty list `silhouette_scores` to store the silhouette scores for different values of k.
silhouette_scores = []

# Iterate over values of k from 2 to 10 (inclusive).
for k in range(2, 20):
    # For each value of k, perform K-means clustering on `df_encoded` and calculate the silhouette score.
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_encoded)
    silhouette_avg = silhouette_score(df_encoded, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Find the index of the maximum silhouette score and add 2 to get the optimal k value.
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2
print(f'The optimal k value is: {optimal_k}')

# Create a line plot of the silhouette scores against the values of k.
# Add x and y axis labels and a title.
# Show the plot.
silhouette_df = pd.DataFrame({'k': range(2, 20), 'Silhouette Score': silhouette_scores})
print(silhouette_df)

alt.Chart(silhouette_df, height=350, width=750).mark_line(point=True).encode(
    x='k:O',
    y='Silhouette Score:Q',
    tooltip=['k', 'Silhouette Score']
).properties(
    title='Silhouette Score vs Number of Clusters (k)'
).interactive()
```

The optimal k value using the Silhouette Coefficient is 2.

Next, we will create clusters using Kmeans and Kmeans++ algorithms with the optimal k value of 2. We will report the performance using the Silhouette Score and compare the results.

```python
from sklearn.cluster import KMeans

# Perform K-means clustering on `df_encoded` with k=2 and random_state=42.
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(df_encoded)

# Calculate the Silhouette Score for K-means and print the result.
kmeans_silhouette = silhouette_score(df_encoded, kmeans_labels)
print(f'Silhouette Score for K-means: {kmeans_silhouette:.4f}')

# Perform K-means++ clustering on `df_encoded` with k=2 and random_state=42.
kmeans_pp = KMeans(n_clusters=2, init='k-means++', random_state=42)
kmeans_pp_labels = kmeans_pp.fit_predict(df_encoded)

# Calculate the Silhouette Score for K-means++ and print the result.
kmeans_pp_silhouette = silhouette_score(df_encoded, kmeans_pp_labels)
print(f'Silhouette Score for K-means++: {kmeans_pp_silhouette:.4f}')

# Compare the Silhouette Scores of K-means and K-means++ and print the algorithm with the better score.
if kmeans_silhouette > kmeans_pp_silhouette:
    print("K-means has a better Silhouette Score.")
elif kmeans_silhouette < kmeans_pp_silhouette:
    print("K-means++ has a better Silhouette Score.")
else:
    print("Both algorithms have the same Silhouette Score.")
```

Both K-means and K-means++ have the same Silhouette Score of 0.5671 when k=2.

Next, we will repeat K-means clustering 50 times, calculate the average Silhouette Score, and compare it with the K-means++ result obtained earlier.

```python
# Initialize an empty list `kmeans_silhouette_scores` to store the Silhouette Scores for each iteration.
kmeans_silhouette_scores = []

# Iterate 50 times:
for _ in range(50):
    # Perform K-means clustering on `df_encoded` with k=2 and random_state=42.
    kmeans = KMeans(n_clusters=2, random_state=42)
    kmeans_labels = kmeans.fit_predict(df_encoded)

    # Calculate the Silhouette Score and append it to `kmeans_silhouette_scores`.
    kmeans_silhouette = silhouette_score(df_encoded, kmeans_labels)
    kmeans_silhouette_scores.append(kmeans_silhouette)

# Calculate and print the average Silhouette Score for K-means over the 50 iterations.
avg_kmeans_silhouette = sum(kmeans_silhouette_scores) / len(kmeans_silhouette_scores)
print(f'Average Silhouette Score for K-means over 50 iterations: {avg_kmeans_silhouette:.4f}')

# Compare the average K-means Silhouette Score with the K-means++ Silhouette Score obtained earlier and print the algorithm with the better score.
if avg_kmeans_silhouette > kmeans_pp_silhouette:
    print("Average K-means has a better Silhouette Score than K-means++.")
elif avg_kmeans_silhouette < kmeans_pp_silhouette:
    print("K-means++ has a better Silhouette Score than average K-means.")
else:
    print("Both algorithms have the same average Silhouette Score.")
```

The average Silhouette Score for K-means over 50 iterations is the same as the Silhouette Score for K-means++ with a single iteration. This suggests that in this specific scenario, the random initialization in K-means does not significantly impact the clustering results compared to the more deterministic initialization of K-means++.

Next, we will apply DBSCAN on the same dataset, find the optimal 'eps' and 'min_samples' values, and compare the number of clusters found with the optimal k value from the Silhouette analysis.

```python
from sklearn.cluster import DBSCAN

# Initialize lists to store silhouette scores and corresponding eps values.
silhouette_scores_dbscan = []
eps_values = []

# Iterate over a range of eps values from 0.1 to 10, with a step of 0.1.
for eps in np.arange(0.1, 10, 0.1):
    # For each eps value, create a DBSCAN model with min_samples=5.
    dbscan = DBSCAN(eps=eps, min_samples=5)

    # Fit the model to the encoded data and get cluster labels.
    dbscan_labels = dbscan.fit_predict(df_encoded)

    # If there's more than one cluster, calculate the silhouette score and store it along with the eps value.
    n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    if n_clusters > 1:
        silhouette_avg = silhouette_score(df_encoded, dbscan_labels)
        silhouette_scores_dbscan.append(silhouette_avg)
        eps_values.append(eps)

# Find the eps value with the highest silhouette score and print it as the optimal eps.
optimal_eps = eps_values[silhouette_scores_dbscan.index(max(silhouette_scores_dbscan))]
print(f'Optimal eps for DBSCAN: {optimal_eps:.2f}')

# Create a DBSCAN model with the optimal eps and min_samples=5.
dbscan_optimal = DBSCAN(eps=optimal_eps, min_samples=5)

# Fit the model to the encoded data and get cluster labels.
dbscan_optimal_labels = dbscan_optimal.fit_predict(df_encoded)

# Calculate and print the number of clusters formed by DBSCAN.
n_clusters_dbscan = len(set(dbscan_optimal_labels)) - (1 if -1 in dbscan_optimal_labels else 0)
print(f'Number of clusters found by DBSCAN: {n_clusters_dbscan}')

# Compare the number of clusters found by DBSCAN with the optimal k value from the Silhouette analysis (which was 2).
if n_clusters_dbscan == optimal_k:
    print("The number of clusters found by DBSCAN is the same as the optimal k value from the Silhouette analysis.")
else:
    print(f"The number of clusters found by DBSCAN ({n_clusters_dbscan}) is different from the optimal k value from the Silhouette analysis ({optimal_k}).")
```

After removing categorical columns and one-hot encoding the data, the optimal number of clusters (k) was determined to be 2 using the Silhouette Coefficient method.

Both K-means and K-means++ clustering algorithms, with k=2, resulted in the same Silhouette Score of 0.5671. This suggests that both methods performed equally well in capturing the underlying structure of the data.

Repeating K-means clustering 50 times did not change the average Silhouette Score, which remained at 0.5671. This indicates that the random initialization of K-means did not significantly affect the clustering results in this case, and the algorithm consistently converged to a similar solution.

DBSCAN, with an optimal eps value of 4.30 and min_samples=5, also identified 2 clusters in the data. This finding aligns with the result from the Silhouette analysis, reinforcing the conclusion that 2 is the appropriate number of clusters for this dataset.

The consistency in the number of clusters identified by both K-means (and its variant K-means++) and DBSCAN suggests a robust underlying structure in the data, with two distinct groups being clearly separable. The similarity in results between these different algorithms further strengthens the validity of this finding.

```python
class Node:
    def __init__(self, key):
        self.node = key
        self.left = None
        self.right = None

def height_tree(A):
    if A is None:
        return 0
    else:
        ldepth = height_tree(A.left)
        rdepth = height_tree(A.right)
        
        if ldepth < rdepth:
            return 1 + ldepth
        else:
            return 1 + rdepth
    

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.left.left.left = Node(7)
root.right.right = Node(6)
root.right.right.right = Node(8)
root.right.right.right.right = Node(9)



print(height_tree(root))
```
